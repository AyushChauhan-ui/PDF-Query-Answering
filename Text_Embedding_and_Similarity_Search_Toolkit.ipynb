```python
import requests
url = "https://arxiv.org/pdf/1708.02709.pdf"


# Download the PDF from the URL
response = requests.get(url)
pdf_content = response.content

# Save the PDF to a file
with open("1708.02709", "wb") as pdf_file:
    pdf_file.write(pdf_content)
```

```python
!pip install PyPDF2
!pip install sentence_transformers
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
import PyPDF2
from langchain.text_splitter import  RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.vectorstores.base import VectorStore
from langchain.vectorstores import FAISS
import os




with open("1708.02709", "rb") as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ""

    for page in pdf_reader.pages:
      text += page.extract_text()

    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len
     )

    ck = text_splitter.split_text(text=text)


#from getpass import getpass
#os.environ[""] = get

embeddings = HuggingFaceEmbeddings()
```

```python
!pip install faiss-cpu

from langchain.vectorstores import FAISS

db = FAISS.from_texts(ck,embedding=embeddings)

print(db)
```

```python
query = """Elaborate Word2Vec"""
#docs = db.similarity_search(query)

docs = db.similarity_search(query=query , k=2)
print(docs)
```
